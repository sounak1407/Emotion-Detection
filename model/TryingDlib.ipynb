```python
import cv2
import numpy as np
import dlib
from skimage.feature import hog
from skimage.transform import resize
from skimage import exposure
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```python
def extract_hog_color(img):
#     rot_img = cv2.rotate(src, cv2.cv2.ROTATE_90_CLOCKWISE)
    img = resize(img, (64,128))
    (fd, hog_image) = hog(img, orientations=9, 
                    pixels_per_cell=(8, 8), cells_per_block=(2, 2), 
                    block_norm='L2-Hys', visualize=True, transform_sqrt=True )
    return hog_image
```

```python
cap = cv2.VideoCapture(0)
while True:
    _,frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces = detector(gray)
    for face in faces:
        x1 = face.left()
        y1 = face.top()
        x2 = face.right()
        y2 = face.bottom()
#         print(face)
        
        #height and width of face
        
        cv2.rectangle(frame, (x1, y1), (x2 , y2),(0,255,0), 2)
        landmarks = predictor(gray,face)
#         print(landmarks)
        
        #General Point
#         x = landmarks.part(27).x
#         y = landmarks.part(27).y
#         cv2.circle(frame,(x,y),3,(255,0,0),-1)
        
        x = x2 - x1
        y = y2 - y1
        
        # Eye Area
        x1_eye = landmarks.part(17).x - int(0.05 * x)
        y1_eye = max(landmarks.part(19).y,landmarks.part(24).y) - int(0.1 * y)
        
        x2_eye = landmarks.part(26).x + int(0.05 * x)
        y2_eye = landmarks.part(45).y + int(0.1 * y)
        
        cv2.rectangle(frame, (x1_eye, y1_eye), (x2_eye , y2_eye),(0,0,255), 2)
        
        eyeArea = gray[y1_eye : y2_eye,x1_eye : x2_eye]
        eyeArea = cv2.resize(eyeArea , (300,150))
        
        #Lips Area
        x1_lips = landmarks.part(48).x - int(0.05 * x)
        y1_lips = landmarks.part(33).y  
        
        x2_lips = landmarks.part(54).x + int(0.05 * x)
        y2_lips = landmarks.part(57).y + int(0.05 * y)
        
        cv2.rectangle(frame, (x1_lips, y1_lips), (x2_lips , y2_lips),(255,0,0), 2)
        
        lipsArea = gray[y1_lips : y2_lips , x1_lips : x2_lips]
        lipsArea = cv2.resize(lipsArea,(300,150))
    
        # hog conversion
        eyeHog = extract_hog_color(eyeArea)
        lipsHog = extract_hog_color(lipsArea)
    
        cv2.imshow("Frame",frame)
        cv2.imshow("Eye Area",eyeArea)
        eyeHog = exposure.rescale_intensity(eyeHog, out_range=(0, 255))
        eyeHog = eyeHog.astype("uint8")
        cv2.imshow("Eye HOG", eyeHog)

        cv2.imshow("Lip Area",lipsArea)
        lipsHog = exposure.rescale_intensity(lipsHog, out_range=(0, 255))
        lipsHog = lipsHog.astype("uint8")
        cv2.imshow("Lips HOG", lipsHog)
        
        final_img = cv2.vconcat([eyeHog, lipsHog])
        cv2.imshow("final IMage",final_img)
        
    
    print(np.array(final_img).shape)
    key = cv2.waitKey(1)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    
cap.release()
cv2.destroyAllWindows()



    
```
